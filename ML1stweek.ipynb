```python

import numpy as np
a = np.array([1,3,4,5,7])
b = np.array([3,9,5,32,5])
a_mean = np.mean(a)
b_mean = np.mean(b)
m = np.sum((a-a_mean)*(b-b_mean))/np.sum((a-a_mean)**2)#m is slope and c is intercept we are finding prediction and precision or error
c = b_mean-m*a_mean
y_precision = m*a+c
error = np.mean((b-b_mean)**2)
print(a)
print(b)
print(m)
print(c)
print(y_precision)
print(error)
print(a_mean)
print(b_mean)
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(np.square(a))
print(np.sqrt(a))
print(np.median(a))
print(np.mode(a))
print(np.median(a))
```

```python
import numpy as np
sales_data = np.array([[200, 230, 260],
                       [180, 220, 150],
                       [240, 258, 230]])
print("The Total Sales Data of First 3 Months.\n", sales_data)

# 2 slicing to analyze data in both types in rows and columns
data_row = sales_data.reshape(1, 9)
print("\nThe Sales Data in one Row to analyze horizontally.\n", data_row)
data_column = sales_data.reshape(9, 1)
print("\nThe Sales Data in one Column to analyze vertically.\n", data_column)

# 3 in This we get data from specific part of data like one month's sales or this product's total sales
month_sale = sales_data[1]
print("\nFirst month and its sales as sample.\n", month_sale)
product_sale = sales_data[:, 0]
print("\nThe first product's sales in all months.\n", product_sale)
spec_product = sales_data[2, 2]
print("\nThe specific product as sample.\n", spec_product)

# 4 we are finding fluctuation rate
noise = np.random.randint(-20, 21, size=sales_data.shape)
fluctuation = sales_data + noise
print("\nThe Fluctuated Sales Rate after (Original sales + noise):\n", fluctuation)
random_sales = np.random.randint(150, 250, (4, 3))
print("\nThe random sales generated for future 4 months.\n", random_sales)

# 5 Find sales average, max sales, and min sales
print("\nThe Average sales of 3 months is: ", np.mean(sales_data))
print("\nThe max sales of product in all months is: ", np.max(sales_data))
print("\nThe min sales of product in all months is: ", np.min(sales_data))

# 6 Revenue
product_prices = np.array([10, 20, 30])
total_revenue = sales_data * product_prices
print("\nThe revenue according to product prices: \n", total_revenue)
sum_product = np.sum(total_revenue, axis=0)
print("\nThe Total revenue of products in all months: ")
print("\n\nProduct 1 = ", sum_product[0])
print("\nProduct 2 = ", sum_product[1])
print("\nProduct 3 = ", sum_product[2])

# 7 Check invertible
determinant = np.linalg.det(total_revenue)
print("\nThe Determinant of Total Revenue is\n", determinant)
if np.linalg.det(total_revenue) != 0:
    inverse = np.linalg.inv(total_revenue)
    print("\nInverse of the total revenue is: \n", inverse)
else:
    print("The total revenue matrix is not invertible")
```
